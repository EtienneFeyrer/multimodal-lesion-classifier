## Extend Datasets with DenseNetPredictions
- extend dev set with out of fold predictions
- extend test set with average fold predictions

```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
drive_path = "/content/drive/MyDrive/ML/"
X_dev_final = "X_dev_final.csv"
X_test_final = "X_test_final.csv"
X_dev = pd.read_csv(drive_path + X_dev_final)
X_test = pd.read_csv(drive_path + X_test_final)
densenet_oof_dev_scores = pd.read_csv("densenet_oof_dev_scores.csv")
#densenet_test_scores = pd.read_csv("densenet_test_scores.csv")
densenet_test_scores = pd.read_csv("densenet_test_scores_full_model.csv")
print(densenet_oof_dev_scores.head())
print(densenet_test_scores.head())
```

```python
# extend the dev and test sets with the densenet scores
X_dev_extended = pd.concat([X_dev, densenet_oof_dev_scores], axis=1)
X_test_extended = pd.concat([X_test, densenet_test_scores], axis=1)
print(X_dev_extended.head())
print(X_test_extended.head())
```

```python
# export the dev and test set as csv files
X_dev_extended.to_csv("X_dev_extended.csv", index=False)
X_test_extended.to_csv("X_test_extended_full_model.csv", index=False)
```
